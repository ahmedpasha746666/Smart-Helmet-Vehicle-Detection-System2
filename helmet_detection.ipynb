```python
from ultralytics import YOLO
```

```python
model = YOLO("yolo11n.pt")  # pretrained small version

# Train without augmentation
model.train(
    data="/content/helmet-and-vehicle-2/data.yaml",          # path to your dataset yaml file
    epochs=50,                # number of training epochs
    imgsz=640,                 # image size
    batch=16,                  # batch size (adjust based on GPU memory)
    workers=4,                 # dataloader workers
    lr0=0.001,                 # initial learning rate
    lrf=0.01,                  # final learning rate fraction
    momentum=0.937,            # optimizer momentum
    weight_decay=0.0005,       # weight decay (L2 regularization)
    warmup_epochs=3,           # warmup epochs
    optimizer="SGD",           # optimizer (SGD or AdamW)
    dropout=0.0,               # no dropout
    mixup=0.0,                 # disable mixup augmentation
    copy_paste=0.0,            # disable copy-paste
    hsv_h=0.0,                 # disable hue augmentation
    hsv_s=0.0,                 # disable saturation augmentation
    hsv_v=0.0,                 # disable brightness augmentation
    flipud=0.0,                # disable vertical flip
    fliplr=0.0,                # disable horizontal flip
    mosaic=0.0,                # disable mosaic
    perspective=0.0,           # disable perspective augmentation
    scale=0.0,                 # disable scaling augmentation
    shear=0.0,                 # disable shearing
    project="runs/train",      # where to save results
    name="yolov11_noaug",      # experiment name
    pretrained=True            # use pretrained backbone
)
```
